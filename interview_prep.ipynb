# Data Structures - Singly Linked List

````python

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
    
    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        last_node = self.head
        while last_node.next:
            last_node = last_node.next
        last_node.next = new_node
    
    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
    
    def delete_value(self, data):
        if not self.head:
            return
        if self.head.data == data:
            self.head = self.head.next
            return
        current_node = self.head
        while current_node.next != None:
            if current_node.next.data == data:
                current_node.next = current_node.next.next
                return
            current_node = current_node.next
    def print_list(self):
        current_node = self.head
        while current_node != None:
            print(f"{current_node.data} --> ", end="")
            current_node = current_node.next
        print()

ll = LinkedList()
ll.append(1)
ll.append(2)
ll.append(5)

ll.print_list()

ll.delete_value(2)

ll.print_list()
````

# Data Structures - Doubly Linked List

````python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.prev = None
        self.next = None
    
class DoublyLinkedList:
    def __init__(self):
        self.head = None
    
    def append(self, data):
        new_node = Node(data)
        if self.head == None:
            self.head = new_node
            return
        last_node = self.head
        
        while last_node.next:
            last_node = last_node.next
        
        last_node.next = new_node
        new_node.prev = last_node

    def prepend(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        new_node.next = self.head
        self.head.prev = new_node
        self.head = new_node
    
    def delete(self, data):
        if not self.head:
            return
        current = self.head
        while current:
            if current.data == data:
                if current.prev:
                    current.prev.next = current.next
                if current.next:
                    current.next.prev = current.prev
                if current == self.head:
                    self.head = current.next
                current = None
                return
            current = current.next


    def print_list(self):
        current = self.head
        while current:
            print(current.data, end=" <-> ")
            current = current.next
        print()

dll = DoublyLinkedList()
dll.append(1)
dll.append(4)
dll.append(6)
dll.print_list()

dll.delete(4)
dll.print_list()

    
````

# Data Structures - Stack (LIFO)

````python
class Stack:
    def __init__(self):
        self.items = []
    
    def is_empty(self):
        return len(self.items) == 0
    
    def push(self, item):
        self.items.append(item)
    
    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            raise IndexError("pop from empty")
    
    def peek(self):
        if not self.is_empty:
            return self.items[-1]
        else:
            raise IndexError("peek from empty")
    
    def size(self):
        return len(self.items)
    
````

# Data Structures - Queue (FIFO)

````python
class Queue:
    def __init__(self):
        self.items = []
    
    def is_empty(self):
        return len(self.items) == 0
    
    def enqueue(self, item):
        self.items.append(item)
    
    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            return None
    
    def peek(self):
        if not self.is_empty():
            return self.items[0]
        else:
            return None
    
    def size(self):
        return len(self.items)
````

````python
class HashMap:
    
    def __init(self, size=10):
        self.size = size
        self.map = [None] * self.size
    
    def _hash(self, key):
        return hash(key) % self.size
    
    def insert(self, key, value):
        index = self._hash(key)
        if not self.map[index]:
            self.map[index] = []
            self.map[index].append((key, value))
        
    def get(self, key):
        index = self._hash(key)
        if self.map[index]:
            for pair in self.map[index]:
                if pair[0] == key:
                    return pair[1]
        return None
    
    def remove(self, key):
        index = self._hash(key)
        if self.map[index]:
            for i, pair in enumerate(self.map[index]):
                if pair[0] == key:
                    del self.map[index][i]
                    return True
        return False
    
    def contains(self, key):
        index = self._hash(key)
        if self.map[index]:
            for pair in self.map[index]:
                if pair[0] == key:
                    return True
        return False
````

# Data Structures - Hash map (open addressing / linear probing)

````python
class HashMap:
    def __init__(self, size=10):
        self.size = size
        self.table = [None] * size

    def _hash(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self._hash(key)

        # Linear probing: look for the next available slot
        original_index = index
        while self.table[index] is not None and self.table[index][0] != key:
            index = (index + 1) % self.size  # Move to the next index

            # If we've looped through the entire table, stop
            if index == original_index:
                raise Exception("HashMap is full")

        self.table[index] = (key, value)

    def get(self, key):
        index = self._hash(key)

        # Linear probing: look for the key in the table
        original_index = index
        while self.table[index] is not None:
            if self.table[index][0] == key:
                return self.table[index][1]
            index = (index + 1) % self.size

            # If we looped back to the original index, the key is not present
            if index == original_index:
                break

        return None

    def delete(self, key):
        index = self._hash(key)

        # Linear probing: look for the key in the table
        original_index = index
        while self.table[index] is not None:
            if self.table[index][0] == key:
                self.table[index] = None
                return
            index = (index + 1) % self.size

            # If we looped back to the original index, the key is not present
            if index == original_index:
                break

        print("Key not found for deletion")

# Example usage
hash_map = HashMap()
hash_map.insert("apple", 1)
hash_map.insert("banana", 2)
hash_map.insert("cherry", 3)

print(hash_map.get("apple"))  # Output: 1
print(hash_map.get("banana"))  # Output: 2
print(hash_map.get("cherry"))  # Output: 3

hash_map.delete("banana")
print(hash_map.get("banana"))  # Output: None (since "banana" was deleted)
````

# Data Structures - Binary search tree

````python
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.value = key

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, key):
        if self.root is None:
            self.root = Node(key)
        else:
            self._insert(self.root, key)

    def _insert(self, root, key):
        if key < root.value:
            if root.left is None:
                root.left = Node(key)
            else:
                self._insert(root.left, key)
        elif key > root.value:
            if root.right is None:
                root.right = Node(key)
            else:
                self._insert(root.right, key)

    def search(self, key):
        return self._search(self.root, key)

    def _search(self, root, key):
        if root is None or root.value == key:
            return root
        if key < root.value:
            return self._search(root.left, key)
        return self._search(root.right, key)

    def delete(self, key):
        self.root = self._delete(self.root, key)

    def _delete(self, root, key):
        if root is None:
            return root

        if key < root.value:
            root.left = self._delete(root.left, key)
        elif key > root.value:
            root.right = self._delete(root.right, key)
        else:
            if root.left is None:
                return root.right
            elif root.right is None:
                return root.left
            temp = self._min_value_node(root.right)
            root.value = temp.value
            root.right = self._delete(root.right, temp.value)
        return root

    def _min_value_node(self, root):
        current = root
        while current.left is not None:
            current = current.left
        return current

    def inorder(self):
        return self._inorder(self.root)

    def _inorder(self, root):
        res = []
        if root:
            res = self._inorder(root.left)
            res.append(root.value)
            res = res + self._inorder(root.right)
        return res
````

````python
class Graph:
    def __init__(self, vertices):
        self.vertices = vertices
        self.matrix = [[0] * vertices for _ in range(vertices)]

    def add_edge(self, u, v):
        self.matrix[u][v] = 1
        self.matrix[v][u] = 1

    def remove_edge(self, u, v):
        self.matrix[u][v] = 0
        self.matrix[v][u] = 0

    def display(self):
        for row in self.matrix:
            print(row)

    def has_edge(self, u, v):
        return self.matrix[u][v] == 1

    def degree(self, v):
        return sum(self.matrix[v])

# Example usage:
g = Graph(5)
g.add_edge(0, 1)
g.add_edge(0, 4)
g.add_edge(1, 2)
g.add_edge(1, 3)
g.add_edge(3, 4)

g.display()
print(g.has_edge(0, 1))
print(g.degree(1))
g.remove_edge(0, 1)
g.display()
````

# Data Structures - Adjacency List Graph

````python
class Graph:
    def __init__(self):
        self.graph = {}

    def add_vertex(self, vertex):
        if vertex not in self.graph:
            self.graph[vertex] = []

    def add_edge(self, vertex1, vertex2):
        if vertex1 in self.graph and vertex2 in self.graph:
            self.graph[vertex1].append(vertex2)
            self.graph[vertex2].append(vertex1)

    def remove_vertex(self, vertex):
        if vertex in self.graph:
            for adjacent in self.graph[vertex]:
                self.graph[adjacent].remove(vertex)
            del self.graph[vertex]

    def remove_edge(self, vertex1, vertex2):
        if vertex1 in self.graph and vertex2 in self.graph:
            if vertex2 in self.graph[vertex1]:
                self.graph[vertex1].remove(vertex2)
            if vertex1 in self.graph[vertex2]:
                self.graph[vertex2].remove(vertex1)

    def get_neighbors(self, vertex):
        if vertex in self.graph:
            return self.graph[vertex]
        return None

    def display(self):
        for vertex, neighbors in self.graph.items():
            print(f"{vertex}: {', '.join(str(neighbor) for neighbor in neighbors)}")

graph = Graph()
graph.add_vertex("A")
graph.add_vertex("B")
graph.add_vertex("C")
graph.add_edge("A", "B")
graph.add_edge("A", "C")
graph.add_edge("B", "C")
graph.display()
graph.remove_edge("A", "B")
graph.display()
graph.remove_vertex("C")
graph.display()
````

# Data Structures - Trie

````python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end_of_word = False


class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end_of_word = True

    def search(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                return False
            node = node.children[char]
        return node.is_end_of_word

    def starts_with(self, prefix):
        node = self.root
        for char in prefix:
            if char not in node.children:
                return False
            node = node.children[char]
        return True

trie = Trie()

trie.insert("apple")
trie.insert("app")
trie.insert("bat")

print(trie.search("apple"))  # Output: True
print(trie.search("app"))    # Output: True
print(trie.search("bat"))    # Output: True
print(trie.search("batman")) # Output: False

print(trie.starts_with("app")) # Output: True
print(trie.starts_with("ba"))  # Output: True
print(trie.starts_with("cat")) # Output: False
````

# Data Structures - Disjoint Set (union-find)

this bascally manges two disjoint or non overlapping sets

it supports two primary operations

find â€”Â meaning it figures out which set that an element belongs to using its parent references

union - it merges the two sets into one 

this is usually used in algorithms that need grouping or partitioning data like Kruskal's algorithm

time complexity of find() is O(alpha(n)) where alpha / Î± is the reverse Ackermann function

time complexity of union() is also O(alpha(n)) the reverse Ackermann function as it does two find operations then performs the union

both operations re very efficient

````python
class DisjointSet:
    def __init__(self, size):
        self.parent = list(range(size))
        self.rank = [0] * size

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, x, y):
        rootX = self.find(x)
        rootY = self.find(y)

        if rootX != rootY:
            if self.rank[rootX] > self.rank[rootY]:
                self.parent[rootY] = rootX
            elif self.rank[rootX] < self.rank[rootY]:
                self.parent[rootX] = rootY
            else:
                self.parent[rootY] = rootX
                self.rank[rootX] += 1

ds = DisjointSet(5)
ds.union(0, 1)
ds.union(1, 2)
ds.union(3, 4)

print(ds.find(0))  # Output: 0
print(ds.find(1))  # Output: 0
print(ds.find(2))  # Output: 0
print(ds.find(3))  # Output: 3
print(ds.find(4))  # Output: 3
````

# Data structures - AVL Tree

an avl tree is essentially a self balancing binary search tree 

however the difference in heights between the left and right subtrees of any node cannot be more than 1

meaning that the "tallness" or height of the left subtree and right subtree cannot have a gap of more than one level

the AVL tree follows these rules
â€”â€”Â binary search tree: for each node, the keys in its left subtree have to be smaller than it, the keys in its right subtree have to be larger than it
â€”â€”Â a node cannot have a left or right subtree that is more than 1 level larger/smaller than the other subtree; either both subtrees are the same depth/height or one of them is just 1 level larger/smaller
â€”â€” if any node has subtrees with a gap in height of more than 1, it is not balanced; perform a left or right rotation restore balance; left rotation is when the right subtree is too tall, right rotation is when the left subtree is too tall

insertion is O(log n), n is total number of nodes in the tree

pre order traversal â€”Â O(n)

space complexity O(n)

the time 

consider this tree

    10
   /  \
  5   20
     /  \
   15   25

the root node 10 has two subtrees

there is the left subtree and the right subtree

the left subtree only has 5, one element, so its height is 1

the right subtree has a height of 2 as it has the level with 20 then a second level with 15 and 25

the difference here is not more than 1

however if the right subtree had a height of 3 then the tree would not be blanaced

then to fix this a rotation would have to happen to make sure the difference is no more than 1

this guarantees the tree is relatively flat, which makes searching and inserting faster

````python
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None
        self.height = 1

class AVLTree:
    def insert(self, root, key):
        if not root:
            return Node(key)
        elif key < root.key:
            root.left = self.insert(root.left, key)
        else:
            root.right = self.insert(root.right, key)

        root.height = 1 + max(self.getHeight(root.left), self.getHeight(root.right))

        balance = self.getBalance(root)

        if balance > 1 and key < root.left.key:
            return self.rightRotate(root)

        if balance < -1 and key > root.right.key:
            return self.leftRotate(root)

        if balance > 1 and key > root.left.key:
            root.left = self.leftRotate(root.left)
            return self.rightRotate(root)

        if balance < -1 and key < root.right.key:
            root.right = self.rightRotate(root.right)
            return self.leftRotate(root)

        return root

    def leftRotate(self, z):
        y = z.right
        T2 = y.left
        y.left = z
        z.right = T2
        z.height = 1 + max(self.getHeight(z.left), self.getHeight(z.right))
        y.height = 1 + max(self.getHeight(y.left), self.getHeight(y.right))
        return y

    def rightRotate(self, z):
        y = z.left
        T3 = y.right
        y.right = z
        z.left = T3
        z.height = 1 + max(self.getHeight(z.left), self.getHeight(z.right))
        y.height = 1 + max(self.getHeight(y.left), self.getHeight(y.right))
        return y

    def getHeight(self, root):
        if not root:
            return 0
        return root.height

    def getBalance(self, root):
        if not root:
            return 0
        return self.getHeight(root.left) - self.getHeight(root.right)

    def preOrder(self, root):
        if not root:
            return
        print(root.key, end=" ")
        self.preOrder(root.left)
        self.preOrder(root.right)

# Example usage
avl = AVLTree()
root = None
keys = [10, 20, 30, 40, 50, 25]

for key in keys:
    root = avl.insert(root, key)

avl.preOrder(root)
````

# Data Structures - red black tree

a red black tree is essentially a binary tree that self-balances

each node has an indicator that says whether it is black or red

the black and red color indicators ensure the tree stays balanced when inserting, deleting

the rules are
â€”â€” every node is either red or black
â€”â€”Â the root node is always black
â€”â€”Â all leaf nodes are black
â€”â€”Â red nodes cannot have red children
â€”â€”Â every path from a node to its descendant NIL nodes has to have the same number of black nodes (height)


````python
class Node:
    def __init__(self, data, color='red'):
        self.data = data
        self.color = color
        self.left = None
        self.right = None
        self.parent = None

class RedBlackTree:
    def __init__(self):
        self.TNULL = Node(0)
        self.TNULL.color = 'black'
        self.root = self.TNULL

    def rotate_left(self, x):
        y = x.right
        x.right = y.left
        if y.left != self.TNULL:
            y.left.parent = x
        y.parent = x.parent
        if x.parent is None:
            self.root = y
        elif x == x.parent.left:
            x.parent.left = y
        else:
            x.parent.right = y
        y.left = x
        x.parent = y

    def rotate_right(self, x):
        y = x.left
        x.left = y.right
        if y.right != self.TNULL:
            y.right.parent = x
        y.parent = x.parent
        if x.parent is None:
            self.root = y
        elif x == x.parent.right:
            x.parent.right = y
        else:
            x.parent.left = y
        y.right = x
        x.parent = y

    def insert(self, key):
        node = Node(key)
        node.parent = None
        node.data = key
        node.left = self.TNULL
        node.right = self.TNULL
        node.color = 'red'

        y = None
        x = self.root
        while x != self.TNULL:
            y = x
            if node.data < x.data:
                x = x.left
            else:
                x = x.right

        node.parent = y
        if y is None:
            self.root = node
        elif node.data < y.data:
            y.left = node
        else:
            y.right = node

        if node.parent is None:
            node.color = 'black'
            return

        if node.parent.parent is None:
            return

        self.fix_insert(node)

    def fix_insert(self, k):
        while k.parent.color == 'red':
            if k.parent == k.parent.parent.right:
                u = k.parent.parent.left
                if u.color == 'red':
                    u.color = 'black'
                    k.parent.color = 'black'
                    k.parent.parent.color = 'red'
                    k = k.parent.parent
                else:
                    if k == k.parent.left:
                        k = k.parent
                        self.rotate_right(k)
                    k.parent.color = 'black'
                    k.parent.parent.color = 'red'
                    self.rotate_left(k.parent.parent)
            else:
                u = k.parent.parent.right
                if u.color == 'red':
                    u.color = 'black'
                    k.parent.color = 'black'
                    k.parent.parent.color = 'red'
                    k = k.parent.parent
                else:
                    if k == k.parent.right:
                        k = k.parent
                        self.rotate_left(k)
                    k.parent.color = 'black'
                    k.parent.parent.color = 'red'
                    self.rotate_right(k.parent.parent)
            if k == self.root:
                break
        self.root.color = 'black'

    def print_tree(self, node):
        if node != self.TNULL:
            self.print_tree(node.left)
            print(f'{node.data} ({node.color})', end=" ")
            self.print_tree(node.right)

    def inorder(self):
        self.print_tree(self.root)
        print()
````

â€”â€” space complexity of O(n)

# Algorithms - Quicksort

````python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)

arr = [3,6,8,10,1,2,1]
sorted_arr = quicksort(arr)
print(sorted_arr)
````

# Algorithms - Merge Sort

````python

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

arr = [38, 27, 43, 3, 9, 82, 10]
sorted_arr = merge_sort(arr)
print(sorted_arr)
````

Time complexity is O(n log n)
because the algorithm divides the array into two halves recursively which takes O(log n) levels of recursion
then at each level the merging function takes O(n) time to combine the two subarrays
so total n * log n = n logn


Space complexity is O(n) because of space needed for temporary arrays while merging subarrays

the recursion stack adds an overhead of log n but the O(n) space for the extra arrays dominates / is more than this, so it is overall O(n) 


# Algorithms - Heap Sort

````python
def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[left] > arr[largest]:
        largest = left

    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heapSort(arr):
    n = len(arr)

    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

    return arr
````

heap sort is a comparison based sorting algorithm

it uses a binary heap to arrange elements in order

if max heap, ensures the maximum element is at the top

if min heap, ensures the minimum element is at the top

the time complexity for building the heap is O(n) and for sorting the array is O(n log n)

so the overall time complexty is O(n log n)

as heapify is called log n times for each element of the n elements

space complexity is O(1), sorting is done in place and no extra space is used other than the input array

# Algorithms - Bubble sort 

````python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

arr = [64, 34, 25, 12, 22, 11, 90]
sorted_arr = bubble_sort(arr)
print(sorted_arr)

````

Bubble sort has a best case time complexity of O(n)

this is if the list is already sorted

because the algorithm will only make one pass through the list and not need any swaps

however worst case time complexity O(n^2) if the list is sorted in reverse order

the average case is also O(n^2) because bubble sort performs a nested loop

the space complexity is O(1) because the sorting is done in place and only needs a constant amount of space

# Algorithms - Binary search (iterative)

````python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else: 
            high = mid -1 
    return -1
````

# Algorithms - Binary search (recursive)

````python
def binary_search(arr, target, low, high):
    if low < high:
        return -1
    
    mid = (low + high) // 2

    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        return binary_search(arr, target, mid + 1, high)
    else:
        return binary_search(arr, target, low, mid - 1)
````

space complexity is O(log n) due to the recursion stack; each split adds a frame to the stack and the recursion depth is logarithmic

but the space complexity is O(1) because it uses a fixed amount of space for variables like low, high, mid, therefore the spce complexity is constant

# Algorithms - Fibonacci dynamic vs recursive

````python
def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)
````

````python
def fibonacci_dynamic(n):
    if n <= 1:
        return n
    fib_cache = [0] * (n + 1)
    fib_cache[1] = 1
    for i in range(2, n + 1):
        fib_cache[i] = fib_cache[i - 1] + fib_cache[i - 2]
    return fib_cache[n]

n = 10
result = fibonacci_dynamic(n)
print(result)
    
````

Time complexity of fibonacci recursive is O(2^n) because for each function call two additional calls are made, so there is an exponential growth

The space complexity of fibonacci recursive is O(n) since the call stack is n levels deep at maximum

Time complexity of fibonacci dynamic is O(n) because the loop is directly from 2 to n, constant time operation 

The space complexity is O(n) since an array fib_cache of size n is used to store the values

# Algorithms - Depth first search

depth first search essentially means traversing a graph by exploring as far as possible along each branch before backtracking

rules
â€”â€” start at a node
â€”â€” visit the node and mark it as visited
â€”â€” recursively visit all unvisited neighbors
â€”â€” backtrack when no unvisited neighbors remain



````python
class Graph:
    def __init__(self):
        self.adjacency_list = {}

    def add_edge(self, start, end):
        if start not in self.adjacency_list:
            self.adjacency_list[start] = []
        if end not in self.adjacency_list:
            self.adjacency_list[end] = []
        self.adjacency_list[start].append(end)

    def depth_first_search(self, start, visited=None):
        if visited is None:
            visited = set()
        visited.add(start)
        print(start, end=" ")  # Process the node (e.g., print it)
        for neighbor in self.adjacency_list.get(start, []):
            if neighbor not in visited:
                self.depth_first_search(neighbor, visited)

    def breadth_first_search(self, start):
        visited = set()
        queue = [start]
        visited.add(start)

        while queue:
            current = queue.pop(0)
            print(current, end=" ")
            for neighbor in self.adjacency_list.get(current, []):
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append(neighbor)

# Testing DFS and BFS
g = Graph()
g.add_edge(1, 2)
g.add_edge(1, 3)
g.add_edge(2, 4)
g.add_edge(3, 5)
g.add_edge(4, 6)
g.add_edge(5, 6)

print("Depth First Search (DFS):")
g.depth_first_search(1)

print("\nBreadth First Search (BFS):")
g.breadth_first_search(1)
````

# Problem Solving - longest substring 

````python
def length_of_longest_substring(s):
    char_set = set()
    left = 0
    max_length = 0

    for right in range(len(s)):
        while s[right] in char_set:
            char_set.remove(s[left])
            left += 1
        char_set.add(s[right])
        max_length = max(max_length, right - left + 1)
    
    return max_length
````

This algorithm uses the sliding window technique to find the longest substring without repeating characters:

A "window" is defined by two pointers: left (start of the substring) and right (end of the substring).

The char_set keeps track of the unique characters currently in the substring.

As the right pointer expands the window by adding new characters:

- If the character is already in the char_set, the left pointer is moved rightward until the repeating character is removed.

- Otherwise, the character is added to the char_set, and the length of the current substring is checked against the max_length.

This ensures we always maintain the longest possible substring without duplicates.


# Problem Solving - Product of array except self

````python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        n = len(nums)
        answer = [1] * n
        
        left_prod = 1
        for i in range(n):
            answer[i] = left_prod
            left_prod *= nums[i]
        
        right_prod = 1
        for i in range(n - 1, -1, -1):
            answer[i] *= right_prod
            right_prod *= nums[i]

        return answer
````

This solution to the Product Except Self problem efficiently calculates the result in  O(n) time without using division. It works by maintaining two running products: one for the left side and one for the right side of each element. First, it initializes an answer array filled with 1s. Then, it computes the left product by iterating through the array and multiplying each element with the cumulative product of all elements to its left, storing this in the answer array. Next, it calculates the right product by iterating backward through the array, multiplying each element in the answer array with the cumulative product of all elements to its right. This way, each position in the answer array ends up containing the product of all elements except the one at that index, combining the left and right products efficiently.

# Problem Solving - palindrome

````python
def is_palindrome(s: str) -> bool:
    left, right = 0, len(s) - 1
    while left < right:
        while left < right and not s[left].isalnum():
            left += 1
        while left < right and not s[right].isalnum():
            right -= 1
        if s[left].lower() != s[right].lower():
            return False
        left += 1
        right -= 1
    return True
````

This function first removes all non-alphanumeric characters and converts the string to lowercase. Then, it checks if the resulting string is equal to its reverse. If they are the same, the string is a palindrome, and the function returns True; otherwise, it returns False.



# Problem Solving - merge interval

````python
def merge(intervals):
    intervals.sort(key=lambda x: x[0])
    result = []
    for interval in intervals:
        if not result or result[-1][1] < interval[0]:
            result.append(interval)
        else:
            result[-1][1] = max(result[-1][1], interval[1])
    return result
````

````python
def solution(A, K):
    # Get the length of the array
    N = len(A)

    # If the array is empty, just return it as is
    if N == 0:
        return A

    # Rotate the array K times
    for _ in range(K):
        # Save the last element
        temp = A[-1]
        
        # Shift all elements to the right by one position
        for j in range(N-1, 0, -1):
            A[j] = A[j-1]
        
        # Place the last element at the beginning
        A[0] = temp

    return A

# Example Usage
A = [5, 6, 7, 8, 4, 2, 7, 1]
K = 2
result = solution(A, K)
print(result)  # Output: [7, 1, 5, 6, 7, 8, 4, 2]
````

        // Calculate the total sum of the array
        int totalSum = 0;
        for (int num : A) {
            totalSum += num;
        }

        int leftSum = 0;
        int rightSum = totalSum;

````python
def solution(K, M, A):
    def can_divide(max_block_sum):
        block_count = 1
        current_sum = 0
        for num in A:
            if current_sum + num > max_block_sum:
                block_count += 1
                current_sum = num
                if block_count > K:
                    return False
            else:
                current_sum += num
        return True
    
    low, high = max(A), sum(A)
    minimal_large_sum = high
    
    while low <= high:
        mid = (low + high) // 2
        if can_divide(mid):
            minimal_large_sum = mid
            high = mid - 1
        else:
            low = mid + 1
    
    return minimal_large_sum
````

        }

    }

# Codility Lesson13 Fibonacci FibFrog

````python
from collections import deque

class Solution:
    def solution(self, A):
        N = len(A)
        
        # Step 1: Create a list of Fibonacci numbers up to N
        fib = [1, 2]
        while fib[-1] <= N:
            fib.append(fib[-1] + fib[-2])
        
        # Step 2: BFS initialization
        queue = deque([(-1, 0)])  # (current position, number of jumps)
        visited = [False] * (N + 1)  # To track visited positions
        visited[-1] = True  # The frog starts at position -1
        
        # Step 3: Perform BFS
        while queue:
            current_pos, jumps = queue.popleft()
            
            # Try all possible Fibonacci distances (jump lengths)
            for f in fib:
                next_pos = current_pos + f
                if next_pos == N:  # If we reach the other bank
                    return jumps + 1
                if next_pos < N and A[next_pos] == 1 and not visited[next_pos]:
                    visited[next_pos] = True
                    queue.append((next_pos, jumps + 1))
        
        return -1  # If we cannot reach the other bank
````

import java.util.*;

public class Solution {
    public int solution(int[] A) {
        int N = A.length;
        
        // Step 1: Generate Fibonacci numbers up to N
        List<Integer> fib = new ArrayList<>();
        fib.add(1);
        fib.add(2);
        while (fib.get(fib.size() - 1) <= N) {
            int nextFib = fib.get(fib.size() - 1) + fib.get(fib.size() - 2);
            fib.add(nextFib);
        }
        
        // Step 2: BFS initialization
        Queue<int[]> queue = new LinkedList<>();
        queue.offer(new int[]{-1, 0});  // (current position, number of jumps)
        boolean[] visited = new boolean[N + 1];
        visited[-1] = true;  // The frog starts at position -1
        
        // Step 3: Perform BFS
        while (!queue.isEmpty()) {
            int[] current = queue.poll();
            int currentPos = current[0];
            int jumps = current[1];
            
            // Try all possible Fibonacci distances (jump lengths)
            for (int f : fib) {
                int nextPos = currentPos + f;
                if (nextPos == N) {  // If we reach the other bank
                    return jumps + 1;
                }
                if (nextPos < N && A[nextPos] == 1 && !visited[nextPos]) {
                    visited[nextPos] = true;
                    queue.offer(new int[]{nextPos, jumps + 1});
                }
            }
        }
        
        return -1;  // If we cannot reach the other bank
    }
}

# Codility Exercise SQL - SQLEventsDelta

````python
WITH RankedEvents AS (
    SELECT 
        event_type, 
        value, 
        time,
        ROW_NUMBER() OVER (PARTITION BY event_type ORDER BY time DESC) AS rn
    FROM events
)
SELECT 
    event_type, 
    (MAX(CASE WHEN rn = 1 THEN value END) - MAX(CASE WHEN rn = 2 THEN value END)) AS value
FROM RankedEvents
GROUP BY event_type
HAVING COUNT(*) > 1
ORDER BY event_type;
````
